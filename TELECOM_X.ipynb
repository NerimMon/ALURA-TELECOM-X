# 📌 Extracción

1. INGRESO DE DATOS

```python
import pandas as pd

archivo='https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
```

```python

datos=pd.read_json(archivo)
datos.head()
```

# 🔧 Transformación

Exploración de Datos

```python


datos
```

```python
datos.info()
```

Normalización de Columnas

```python
datos_list = list()

#desde la columna 2 en adelante normalizo los datos para cada columna creando una lista
for col in datos.columns[2:]:
    v = pd.json_normalize(datos[col])
    v.columns = [f'{col}_{c}' for c in v.columns]
    datos_list.append(v)

# combino todo en un dataframe
datos_normalized = pd.concat([datos.iloc[:, :2]] + datos_list, axis=1)
```

```python
datos_normalized.head(3)
```

Revisión de Inconsistencias

```python
columnas=list(datos_normalized.columns)
```

```python
for c in columnas:
  print(f'{len(pd.unique(datos_normalized[c]))} Datos unicos para {c}:  {pd.unique(datos_normalized[c])}')
```

Cambio de Valores

```python
datos_normalized['Churn']=datos_normalized['Churn'].replace('','No')
datos_normalized['phone_MultipleLines']=datos_normalized['phone_MultipleLines'].replace('No phone service','No')
datos_normalized['internet_InternetService']=datos_normalized['internet_InternetService'].replace('DSL','Yes').replace('Fiber optic','Yes')
datos_normalized=datos_normalized.replace('No internet service','No')
datos_normalized['account_PaymentMethod']=datos_normalized['account_PaymentMethod'].str.replace(' (automatic)','')
```

```python
for c in columnas:
  print(f'{len(pd.unique(datos_normalized[c]))} Datos unicos para {c}:  {pd.unique(datos_normalized[c])}')
```

```python
import matplotlib.pyplot as plt

# Activar estilo moderno
plt.style.use('seaborn-v0_8-whitegrid')

# Crear figura y ejes
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
plt.subplots_adjust(top=0.8, left=0.15, wspace=0.6)
fig.suptitle('Análisis de renuncia de clientes', fontsize=20, fontweight='bold')

# Datos (mismos que en tu código)
labels = ['Permanecen', 'Renunciaron']
values = churn.values
colors_bar = ['#3B5B92', '#FF7F50']  # azul oscuro y naranja coral

# --- Gráfico de Barras ---
ax[0].bar(labels, values, color=colors_bar, edgecolor='none')
ax[0].set_title('Renuncias en cantidad', fontsize=16, pad=20)
ax[0].set_ylabel('Cantidad de clientes', fontsize=12)
ax[0].set_frame_on(False)

# Etiquetas sobre las barras
for i, v in enumerate(values):
    ax[0].text(i, v + max(values)*0.02, str(v), ha='center', fontsize=12, fontweight='bold')

# --- Gráfico Circular ---
ax[1].pie(values,
          labels=labels,
          autopct='%1.1f%%',
          startangle=90,
          colors=colors_bar,
          wedgeprops={'edgecolor': 'white', 'linewidth': 1})
ax[1].set_title('Porcentajes de Churn de clientes', fontsize=16, pad=20)
ax[1].axis('equal')

# Guardar y mostrar
fig.savefig('Grafico_Churn01.png', dpi=300, bbox_inches='tight')
plt.show()
```

CORRELACIÓN

```python
df_co=datos_normalized.drop(['customerID'],axis=1)
df_co.head(3)


df_corr = pd.get_dummies(df_co, columns=['Genero','Contrato','Metodo_pago'], drop_first=True)
df_corr.head(3)
```

VARIABLES TOTALES

```python
df_correlacion=df_corr.corr()
df_correlacion
```

```python
plt.figure(figsize=(15,10))

# Heatmap con gama de morado más específica
sns.heatmap(
    df_correlacion.corr(numeric_only=False),
    annot=True,
    annot_kws={'size': 10},  # fuente un poco más grande
    cmap="Purples",          # gama de morados
    linewidths=0.5,          # separadores entre celdas
    linecolor='gray',        # color de los separadores
    vmin=-1, vmax=1          # rango de correlación
)

plt.title("Correlación entre variables", fontsize=16, weight='bold', ha='center')
plt.savefig('Grafico_Correlacion01.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

CUENTA DIARIA CONTRA EVASION

```python
a=df_correlacion['Churn'].corr(df_correlacion['Cuentas_diarias'])
print(f'La CORRELACIÓN entre EVASIÓN y CUENTAS DIARIAS es de {round(a,2)}')
```

INTRODUCCION

Telecom X es una empresa de telecomunicaciones que enfrenta un alto índice de evasión de clientes (Churn). Como analista de datos, el objetivo principal es identificar patrones y factores que influyen en la renuncia de clientes, con el fin de proponer estrategias de retención. Para esto se realizó un análisis de datos históricos de clientes, incluyendo información demográfica, tipo de contrato, servicios contratados y facturación. El análisis incluyó desde la extracción y transformación de datos hasta la visualización y correlación de variables para comprender mejor los factores asociados al Churn.

1. Extracción y carga de datos

* Qué se hizo: Se cargó el archivo JSON de los clientes en un DataFrame de pandas.
* Objetivo: Tener los datos listos para análisis y transformación.
* Resultado: Se identificaron las columnas y tipos de datos disponibles, incluyendo información demográfica, servicios contratados, facturación y estado de Churn.

2. Transformación de datos

**¿Qué se hizo?**

* Normalización de los datos (convertir estructuras anidadas en columnas planas).
* Separación de columnas numéricas y de texto para análisis específico.

**Objetivo:** Preparar los datos para cálculos estadísticos y visualizaciones.

**Resultado:** Dos conjuntos: datos_normalized para análisis general y datos_texto para variables categóricas como género, contrato y método de pago.

3. Análisis descriptivo

**¿Qué se hizo?**

* Se utilizó .describe() para variables numéricas y categóricas.
* Se observó distribución de facturación mensual, total gastado y meses de contrato.

**Objetivo:** Conocer rangos, medias, máximos y mínimos de los datos.

**Resultado:** Primer panorama de clientes que permanecen vs. renuncian.

4. Visualización del Churn

**¿Qué se hizo?**

* Gráficos de barras y pastel para mostrar la distribución de clientes que permanecen o renuncian.

* Boxplots para analizar relación del Churn con facturación mensual, gasto total y meses de contrato.

**Objetivo:** Visualizar claramente quiénes están renunciando y posibles patrones.

**Resultado:**

* La mayoría de los clientes permanecen, pero existe un grupo significativo que renuncia.

* Clientes que renuncian tienden a tener menor número de meses de contrato y facturación más baja en promedio.

5. Análisis de variables categóricas

**¿Qué se hizo?**

* Se examinaron columnas como Genero, Contrato y Metodo_pago.

**Objetivo:** Detectar patrones de Churn en distintos grupos de clientes.

**Resultado:** Posibles diferencias entre tipos de contrato y método de pago en relación al Churn.

6. Correlación entre variables

**¿Qué se hizo?**

* Se calculó la matriz de correlación entre variables numéricas y codificadas (get_dummies) para identificar relaciones.

* Se graficó un heatmap para visualizar la fuerza de la correlación.

**Objetivo:** Determinar qué variables están más asociadas a la renuncia de clientes.

**Resultado:**
* La facturación mensual, el total de gasto y el número de servicios están relacionados con el Churn.

* Correlación con cuentas diarias: 0.26 (positiva, débil).

* Correlación con número de servicios: -0.6 (negativa, fuerte), indicando que cuantos más servicios tiene un cliente, menos probabilidad de renunciar.

7. Servicios contratados y número de servicios

**¿Qué se hizo?**

* Se sumaron todos los servicios contratados por cada cliente para crear Numero_de_Servicios.

* Se evaluó la correlación de este número con Churn.

**Objetivo:** Analizar si la cantidad de servicios contratados influye en la evasión.

**Resultado:** Clientes con más servicios tienden a permanecer, mientras que quienes renuncian suelen tener menos servicios.

8. Conclusión general

El Churn está asociado principalmente a:

* Facturación mensual y total: Clientes con menor gasto tienden a renunciar más.

* Número de servicios contratados: Mayor número de servicios reduce el riesgo de Churn.

* Meses de contrato: Contratos más cortos presentan mayor evasión.

Estas conclusiones ayudan a identificar clientes en riesgo y generar estrategias de retención, como ofrecer paquetes de servicios adicionales o incentivos a clientes con contratos cortos.
